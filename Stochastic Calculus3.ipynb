 ## 블랙숄즈 Partial Differential Equation
 
 내 포트폴리오 가치를 $\pi$라고 가정하자.
 
 포트폴리오는 파생상품 1개 매수와 주식 $\Delta$ 개 매도 포지션이고 파생상품의 가치는 F이다.
 
 파생상품의 가치는 시간과 기초자산에 의존한다. F = F(t,S)
 
 $ \pi = F - \Delta \bullet S $
 
 $ dS = \mu S dt + \sigma S dW $
 
 1)포트폴리오 가치변화
 $$
 d\pi = dF - \Delta dS
 $$
 
 
 2)이토공식에 의해
 $$
 dF = F_t dt + F_S dS + \frac{1}{2}F_{SS} {(dS)}^2 = F_t dt + F_S (\mu S dt + \sigma S dW ) + \frac{F_{SS} \sigma^2 S^2}{2}dt \\
 \ \\
 dF = (F_t + \mu S F_S + F_{SS} \frac{\sigma^2 S^2 }{2} )dt + \sigma S \bullet F_{S} dW
$$

2)를 1)에 대입

$$
d\pi = (F_t + \mu S F_S + F_{SS} \frac{\sigma^2 S^2 }{2} )dt + \sigma S \bullet F_{S} dW - \Delta (\mu S dt + \sigma S dW) 
\\ \ \\
 = (F_t + \mu S F_S - \Delta \mu S  + F_{SS} \frac{\sigma^2 S^2 }{2} )dt + (\sigma S F_{S} -  \Delta \sigma S )dW
$$

만약 기존 포트폴리오가 Perfect Hedge Portfolio라면 Non-Stochastic Process이여야 한다. 즉 dW term 의 계수가 0이 되어야한다.

$$
F_S = \frac{\delta F}{\delta S} = \Delta
$$

따라서, 기존 포트폴리오가 Perfect Hedge Portfolio라면 최초 포트폴리오는 다음과 같다.
$ \pi = F - F_S \bullet S $

3) 또한, 2)에서 $F_S = \Delta$ 에 따라 나머지 term이 지워지므로 포트폴리오 가치변화는 다음과 같다.
$$
d\pi = (F_t + F_{SS} \frac{\sigma^2 S^2 }{2} )dt
$$

4) 기존 포트폴리오가 무위험이므로 포트폴리오의 가치변화는 Risk Free 만큼 증가한다.
$$
d\pi = \pi \times r dt = (F - F_S \bullet S) \bullet r dt
$$

3)과 4)에 의해
$$
(F_t + F_{SS} \frac{\sigma^2 S^2 }{2} )dt = (F - F_S \bullet S) \bullet r dt
$$

어떤 파생상품과 기초자산의 관계는 다음과 같다. (옵션, 선물 관계없이)

$
F_t + rSF_S + F_{SS} \frac{\sigma^2 S^2 }{2} = rF
$

이를 블랙숄즈 PDE라고한다. 

이를 차분방정식으로 바꾸어 Numerical Solution을 구한 것을 Finite Differece Method 라고 한다.

$$
\frac{F(t+\Delta t) - F(t)}{\Delta t} + rS \bullet \frac{F(S+\Delta S) - F(S)}{\Delta S} + \frac{\sigma^2 S^2}{2}\frac{F(S + \Delta S) - 2 F(S) + \Delta (S - \Delta S)}{{(\Delta S)}^2} = rF
\\ \ \\ 
F(t+\Delta t) = rF \Delta t + F(t) - rS \Delta t \bullet \frac{F(S+\Delta S) - F(S)}{\Delta S} - \Delta t \frac{\sigma^2 S^2}{2}\frac{F(S + \Delta S) - 2 F(S) + \Delta (S - \Delta S)}{{(\Delta S)}^2}
$$

FDM에 대해서는 다른 챕터에서 설명한다.
